```python
from datetime import datetime, timedelta

today = datetime(2025, 3, 27, 8, 34) 
for i in range(4980): 
    date_str = df.loc[i, "date_of_experience"]

    if "hour" in date_str:
        hours = int(date_str.split()[0])
        df.loc[i, "date_of_experience"] = (today - timedelta(hours=hours)).strftime("%b %d, %Y")
        
    elif "day" in date_str:
        days = int(date_str.split()[0]) if date_str.split()[0].isdigit() else 1
        df.loc[i, "date_of_experience"] = (today - timedelta(days=days)).strftime("%b %d, %Y")
        
    elif "Updated" in date_str:
        date_str = date_str.replace("Updated ", "")
        df.loc[i, "date_of_experience"] = date_str
```

# Analyzing  the sentiment of customer_review (positive , negative, neutral)

```python
from textblob import TextBlob
```

```python
def analyze_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity 
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
        
df["sentiment"] = df["customer_review"].apply(analyze_sentiment)

print("Sentiment Distribution:")
print(df["sentiment"].value_counts())
```

# Analyzing the sentiment of customer_review (positive , negative, neutral) for 2024

```python
df_2024 = df[df["date_of_review"].dt.year == 2024]

def analyze_sentiment(row):
    if row["rating_out_of_5"] in [4, 5]:
        return "Positive"
    elif row["rating_out_of_5"] == 3:
        return "Neutral"
    elif row["rating_out_of_5"] in [1, 2]:
        return "Negative"
    else:
        polarity = TextBlob(str(row["customer_review"])).sentiment.polarity
        if polarity > 0:
            return "Positive"
        elif polarity < 0:
            return "Negative"
        else:
            return "Neutral"

df_2024["sentiment"] = df_2024.apply(analyze_sentiment, axis=1)
print("Sentiment Distribution for 2024:")
print(df_2024["sentiment"].value_counts())
```

```python
def update_sentiment(row):
    if row["rating_out_of_5"] in [4.0, 5.0]:
        return "Positive"
    elif row["rating_out_of_5"] == 3.0:
        return "Neutral"
    elif row["rating_out_of_5"] in [1.0, 2.0]:
        return "Negative"
    else:
        return row["sentiment"]  # Keep the existing sentiment value if no rating matches

# Update the sentiment column based on star ratings
df_2024["sentiment"] = df_2024.apply(update_sentiment, axis=1)

# Print sentiment distribution after the update
print("Updated Sentiment Distribution for 2024:")
print(df_2024["sentiment"].value_counts())
```
